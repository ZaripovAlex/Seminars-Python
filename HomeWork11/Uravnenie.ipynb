# f(x) =0.6x^3+5.5x^2+10x -5

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
from sympy import *

a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol("x")

f = a*(x**3)+b*(x**2)+c*(x*1)+d
```

```python
f
```

# Строим график 

```python
from sympy.plotting import plot

grafik1 = plot(f,(x,-20,20), line_color='red', title='график функции')
```

# Находим корни уровнения

```python
print("Корень 1", solve(f,x)[0].n())
print("Корень 2", solve(f,x)[1].n())
print("Корень 3", solve(f,x)[2].n())
```

# Производная

```python
dif = diff(f)
dif
```

```python
grafik2 = plot(dif,(x,-20,20), line_color='red', title='график производной')
```

# Совмещаем графики

```python
sovm_grafik = plot(f, dif,(x,-10,10), title="Совмещенный график", legend= True, xlabel='X', ylabel='Y', show=False)
sovm_grafik[0].line_color = 'red'
sovm_grafik[1].line_color = 'green'

sovm_grafik.show()
```

# Определяем вершины

## Корни производной¶

```python
print("Корень 1", solve(dif,x)[0].n(3))
print("Корень 2", solve(dif,x)[1].n(3))
```

# Интервалы на которых функция возрастает

```python
 solve(dif > 0,x)
```

# Интервалы на которых функция убывает

```python
 solve(dif < 0,x)
```

# Промежутки, на котором f > 0

```python
solve(f > 0,x)
```

# Промежутки, на котором f < 0

```python
solve(f < 0,x)
```
